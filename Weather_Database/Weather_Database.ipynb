```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Create a set of 2,000 random latitudes and longitudes

```python
# Create a nandom set Latitude from -90 to 90 deg
lats = np.random.uniform(low = -90.000, high = 90.000, size = 2000)
# Create a random set of longitude from -180 to + 180 deg
lngs = np.random.uniform(low = -180.000, high = 180.000, size = 2000)
lat_lngs = zip(lats, lngs)

# Save the latitude and longitudes in a coordinate list.
coordinates = list(lat_lngs)
```

### Get the Nearest City from ciipy

```python
# Import citipy module to get the nearest city based on latitude and longitude.
from citipy import citipy

# Create a list for holding city names.
cities = []

# Identify the nearest city for each latitude and longitude combination
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    # If the city is unique, then add the name in the cities list.
    if city not in cities:
        cities.append(city)
# Check if there are sufficient number of cities
len(cities)
```

### Perform an API call with the OpenWeatherMap

### Build empty cell to hold data and construct the data retrieval

```python
# Import requests and API keys to request the API
import sys
sys.path.append('..')

import requests
from config import weather_api_key

# Import the datetime library to convert the time stamp
from datetime import datetime

# API call from OpenWeatherMap url
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID="+weather_api_key
```

```python
# Create an empty list to hold the weather data.
city_data = []

# Print the beginning of the logging of city names.
print("Beginning Data Retrieval         ")
print ("---------------------------------")

# Create counters.
record_count = 1
set_count = 1
```

```python
# Loop through all the cities in the list. 
for i, city in enumerate(cities):
    
    #Group cities in set of 50 for logging purposes.
    if (i %  50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    # Create endpoint URL with each city.
    city_url = url + "&q=" + city.replace(" ", "+")
    
    # Log the URL, record, and set numbers and the city. 
    print(f"Processing Record{record_count} of Set {set_count} | {city}")
    # Add 1 to the record count.
    record_count += 1
    
# Run an API request for each of the cities:
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        # Parse out the needed data. 
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_current = city_weather["weather"][0]["description"]
                
        # Append the city information into city_data list. 
        city_data.append({"City": city.title(),
                          "Country": city_country,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                           "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Current Description": city_current})

# If an error is experienced, skip the city.
    except:
        print("City not found. Skipping ...")
        pass
# Indicate that the Data loading s complete.
print("----------------------------------")
print("Data Retrieval Complete")
print("----------------------------------")
         
```

```python
### Create a DataFrame from the city_data dictionary
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

### Create the Output as a CSV File from the DataFrame.

```python
# Create the outup file (CSV)
output_data_file =  "WeatherPy_Database.csv"

# Export the output file into a CSV file.
city_data_df.to_csv(output_data_file, index_label = "City_ID")
```
